```python
import os
import pandas as pd

# Set the directory path where the CSV files are stored
directory_path = '/Users/shweta/Downloads/raw_csv'

# Create an empty list to store the common column names
common_columns = []

# Loop through all the CSV files in the directory
for file_name in os.listdir(directory_path):
    if file_name.endswith('.csv'):
        # Read the CSV file and get its column names
        column_names = set(pd.read_csv(os.path.join(directory_path, file_name),encoding='ISO-8859-1').columns)
        # Append the set of column names to the list of common column names
        common_columns.append(column_names)

# Find the intersection of all the sets of column names
common_column_names = list(set.intersection(*common_columns))

print('Common column names:', common_column_names)
```

```python





import pandas as pd
import os

# Define the path of the directory containing the CSV files
directory_path = '/Users/shweta/Downloads/raw_csv'

# Define the common column name
common_column_name = 'SEQN'

# Create an empty list to store the dataframes
df_list = []

# Loop through the files in the directory
for filename in os.listdir(directory_path):
    if filename.endswith('.csv'):
        # Read the CSV file into a dataframe
        filepath = os.path.join(directory_path, filename)
        df = pd.read_csv(filepath, encoding='ISO-8859-1')
        # Append the dataframe to the list
        df_list.append(df)

# Merge the dataframes on the common column
merged_df = pd.concat(df_list)
final_df = merged_df.merge(df_list[0], on=common_column_name)

# Save the merged dataframe to a new CSV file
final_df.to_csv('merged_file_new.csv', index=False)
```
