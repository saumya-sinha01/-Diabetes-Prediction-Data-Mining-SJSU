```python
categorical_col_names = []
num_col_names = []
count=0
for col in df.columns:
  if df[col].dtype == 'object':
    categorical_col_names.append(col)
    
  else:
    num_col_names.append(col)
    count+=1

print("categorical feature list : ", categorical_col_names)
print("")
print("Numerical feature list : ", num_col_names)
print("")
print(count)
```

```python
for col in num_col_names:
  print(col, '({})'.format(df[col].nunique()), ":")
  print(df[col].value_counts())
  print()
  print()
```
